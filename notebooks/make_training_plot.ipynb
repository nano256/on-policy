```python
def get_exp_name(file_path):
    start_idx = len(exp_path) + 1
    end_idx = str(file_path).find('/', start_idx)
    return str(file_path)[start_idx:end_idx]

def get_run_number(file_path):
    substr = 'run'
    idx = str(file_path).find(substr) + len(substr)
    return str(file_path)[idx:idx+1]

def get_seed(file_path):
    substr = 'seed_'
    start_idx = str(file_path).find(substr) + len(substr)
    end_idx = str(file_path).find('_', start_idx)
    return int(str(file_path)[start_idx: end_idx])

def get_glob_seed_pattern(file_path):
    substr = 'seed_'
    start_idx = str(file_path).find(substr) + len(substr)
    end_idx = str(file_path).find('_', start_idx)
    return str(file_path)[:start_idx] + '?' + str(file_path)[end_idx:]

def clean_up_dict(data):
    new_data = {}
    for k, v in data.items():
        new_k = k.split('/')[-1]
        new_data[new_k] = v
    return new_data

```

```python
def get_stat_data(file_data):
    stat_data = {}

    for exp_name, exp_data in file_data.items():
        if exp_name not in stat_data.keys():
            stat_data[exp_name] = {}
        raw_data = {}
        for seed, seed_data in exp_data.items():
            for value_name, values in seed_data.items():
                if value_name not in raw_data.keys():
                    raw_data[value_name] = []
                raw_data[value_name].append(values)
        for value_name, values in raw_data.items():
            values = np.array(values)
            if value_name not in stat_data.keys():
                stat_data[exp_name][value_name] = {}
            stat_data[exp_name][value_name]['mean'] = np.mean(values[..., 2], axis=0)
            stat_data[exp_name][value_name]['std'] = np.std(values[..., 2], axis=0)
            for timesteps in values[..., 1]:
                assert np.all(values[0,:,1] == timesteps)
            stat_data[exp_name][value_name]['timestep'] = values[0, :, 1]
    return stat_data
```
